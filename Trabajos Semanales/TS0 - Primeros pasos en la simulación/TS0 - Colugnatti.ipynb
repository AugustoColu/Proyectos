## Introducción
 El objetivo de esta TS0 fue diseñar un generador de señales con el lenguaje de programación Python. La idea es que estas señales generadas de manera digital con las librerías Numpy y Scipy "simulen", en cierto sentido, las lecturas que toma un ADC (*Analog to Digital Converter*) sobre una señal analógica. Es importante destacar que para muestrear una señal se deben establecer dos parámetros fundamentales: la cantidad de muestras que se deseen obtener y la frecuencia con que se tomen (*frecuencia de muestreo*), es decir, la cantidad de muestras que toma el ADC por segundo. Por otro lado, las ondas sinusoidales con las que vamos a trabajar tienen la siguiente pinta $y(t) = A_{max} . \sin (2\pi . f . t + \phi) + V_{medio}$


## Desarrollo práctico

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sp

def func_senoidal (a_max, frec, fase, cant_muestras, frec_muestreo, v_medio):
    
    Ts = 1/frec_muestreo
    t_final = cant_muestras * Ts
    tt = np.arange (0, t_final, Ts) # defino una sucesión de valores para el tiempo
    xx = a_max * np.sin (2 * np.pi * frec * tt + fase) + v_medio # tt es un vector, por ende la función sin se evalúa para cada punto del mismo
    # xx tendrá la misma dimensión que tt
    
    return tt, xx

### Inicializo variables en cero ###

a_max = 0
frec = 0
fase = 0
cant_muestras = 0
frec_muestreo = 0
v_medio = 0
```

```python
### Señal 1 ###

tt_1, ss_1 = func_senoidal (a_max = 1, frec = 10, fase = 0, cant_muestras = 1000, frec_muestreo = 1000, v_medio = 0)

plt.subplot (6, 1, 1)
plt.plot (tt_1, ss_1, linestyle='-', color='black')
plt.title ("Onda senoidal de 10 Hz")
plt.xlabel ("Tiempo")
plt.ylabel ("Volts")
plt.grid (True)
plt.show ()
```
